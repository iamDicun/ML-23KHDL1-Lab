```python
import pandas as pd
import os
```

## Read file csv and summary

```python
df = pd.read_csv('hotel_urls.csv')

print("Data shape: ", df.shape)

print("5 first rows")
display(df.head(5))

print("5 last rows")
display(df.tail(5))

display(df.info())
```

## Check duplicate

```python
print("The number of duplicate: ", df.duplicated().sum())
```

## Check missing data

```python
print(df.isnull().sum())
```

- Given the large number of hotels and the importance of the missing attribute, we have decided to delete the rows with missing data.

```python
df = df.dropna()

print("Data shape after deletion", df.shape)
```

## Filter hotel

```python
print("Number of hotels with at least 100 reviews: ", len(df[df['review_count'] >= 100]))
```

- We assume that hotels with at least 100 reviews provide a more diverse and comprehensive range of sentiments and aspects. Given the substantial number of hotels meeting this criteria, we have decided to exclude those with fewer than 100 reviews to ensure data quality.

```python
df = df.drop(df[df['review_count'] < 100].index)

df = df.drop(df[df['review_count'] > 1000].index)

print("Data shape after deletion", df.shape)
```

## Export file csv

```python
os.makedirs("../processed_data", exist_ok=True)
df.to_csv("../processed_data/hotel_urls.csv", index=False, encoding='utf-8-sig')

print("File saved successfully")
```
